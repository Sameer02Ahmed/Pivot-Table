Insert a pivot table and show "Gross income" for a specific branch

```python
pivot_table = pd.pivot_table(table, values='gross income', index='Branch', aggfunc='sum', margins=True)
```

```python
pivot_table
```

In Pivot table, show gross income according to the "Gender" in specific branch

```python
table1 = pd.pivot_table(table, values='gross income', index='Branch', columns='Gender', aggfunc='sum', margins=True)
```

```python
table1
```

In Pivot table, show gross income according to the "Customer type" in specific branch

```python
table2 = pd.pivot_table(table, values=['gross income'], index=['Customer type','Branch'], columns=['Gender'],aggfunc='sum', margins=True)
```

```python
table2
```

In the Pivot table, show the gross income according to the "Payment" in specific branch
